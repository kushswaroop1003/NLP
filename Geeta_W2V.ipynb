```python
text = extract_text("/content/Bhagavad_Gita_As_It_Is.pdf")
print(text)
```

```python
pattern = re.compile(r"[a-zA-Z]+,{1}\s{1}")
```

```python
matches = pattern.findall(text)
print(matches)
```

```python
import nltk
nltk.download('punkt')
```

```python
sentence = nltk.sent_tokenize(text)
sentence
```

```python
paragraph = []
for i in range(len(sentence)):
  sent = re.sub(r'^[a-zA-Z]'," ",sentence[i])
  my_word = nltk.word_tokenize(sent)
  paragraph.append(my_word)
```

```python
import gensim
from gensim.models import Word2Vec,KeyedVectors
```

```python
!pip install wget
```

```python
model = gensim.models.Word2Vec(window =10, min_count=2)
```

```python
model.build_vocab(paragraph)
```

```python
model.train(paragraph, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.wv.most_similar('Arjuna')
```

```python
model.wv.most_similar('Krishna')
```

```python
model.wv.most_similar('Duryodhana')
```

```python
model.wv.most_similar('sanätana-dharma')
```

```python
model.wv.most_similar('stress')
```
